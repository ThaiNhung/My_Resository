```python
#Grayscale convert
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#convert the color image to grayscale image in the imread function by add the parameter is 0 (CV_LOAD_IMAGE_GRAYSCALE)
img = cv2.imread('C:/Data/Data/IMG/grayImage.JPG',0)
#img = imread("img_sample.png", CV_LOAD_IMAGE_GRAYSCALE);
plt.imshow(img,cmap='gray')
```

```python
# # case 2 - Use the cvtColor command
#cvtColor(source, dest, mode) allow convert the color image "source" to grayscale image and save to "dest" with mode "CV_BGR2GRAY" allow convert color image and save with format BGR to grayscale image.
img = cv2.imread('C:/Data/Data/IMG/grayImage.JPG')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap = 'gray')
```

```python
def show_pic(img):
    #Increase image
    fig = plt.figure(figsize=(15,15))
    #add_subplot(111)equal add_subplot(1, 1, 1) - "111" mean is "grid 1x1, first box" and "234" mean "grid 2x3, box 4".
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap ='gray')
```

```python
show_pic(img)
```

```python
#Black-and-white convert
img = cv2.imread('C:/Data/Data/IMG/grayImage.JPG',0)
#ret, th1 =cv2.threshold(img, 127, 255,cv2.THRESH_BINARY)
ret, th1 =cv2.threshold(img, 127, 255,0)
plt.imshow(th1)
```

```python
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8)
show_pic(th2)
```

```python
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY_INV,11,8)
show_pic(th2)
```

```python
#Gradients - Gradients of light intensity
img2= cv2.imread('C:/Data/Sudoku.png',0)
plt.imshow(img2)
```

```python
def display_img(img2):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img2,cmap ='gray')
```

```python
#the first - we need calculate the sobelX
sobelX = cv2.Sobel(img2,cv2.CV_64F,1,0,ksize=5)
```

```python
display_img(sobelX)
```

```python
#the second - we need calculate the sobelY
sobelY = cv2.Sobel(img2,cv2.CV_64F,0,1,ksize=5)
display_img(sobelY)
```

```python
laplacian = cv2.Laplacian(img2, cv2.CV_64F,ksize = 1,scale =1,delta=0)
#Laplacian(const Mat& src, Mat& dst, int ddepth, int ksize=1, double scale=1, double delta=0, int borderType)
display_img(laplacian)
```
